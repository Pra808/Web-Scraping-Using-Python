This is a project of web scrapping of world ICC test men Ranking

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

Using requests library, we can fetch the content from the URL given and beautiful soup library helps to parse,fetch the details the way we want. Use requests and Beautiful Soup for scraping and parsing data from the Web. You can use a beautiful soup library to fetch data using Html tag, class, id, css selector and many more ways. Pandas makes it easy to scrape a table (

tag) on a web page. After obtaining it as a DataFrame, it is of course possible to do various processing and save it as an Excel file or csv file.


```python
url="https://www.cricbuzz.com/cricket-stats/icc-rankings/men/batting"
```

Above is the link of the website which we are going to scrap.

```python
page=requests.get(url)
page
```

we request the website once we get the response it means that we can scrap the website. Basically we are asking  for the permission from the website.

```python
data=page.text
```

```python
s=bs(data,"html.parser")
```

```python
print(s.prettify())
```

Beautiful Soup is a Python package for parsing HTML and XML documents (including having malformed markup, i.e. non-closed tags, so named after tag soup). It creates a parse tree for parsed pages that can be used to extract data from HTML, which is useful for web scraping.
The prettify() method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string, with a separate line for each tag and each string:

```python
title=s.title
print(title.text)
```

```python
player=s.find_all("a",class_="text-hvr-underline text-bold cb-font-16")
for i in range(0,10):
    print(player[i].text)

```

by using soup.findall() we can find the given tag by using it class or id we can find the information what we want  it and we will convert it in text

```python
country=s.find_all("div",class_="cb-font-12 text-gray")
for i in range(0,10):
       print(country[i].text)
```

by using soup.findall() we have scrap the all div tag having the following class

```python
position=s.find_all("div",class_="cb-col cb-col-16 cb-rank-tbl cb-font-16")
for i in range(0,10):
  print(position[i].text)
```

```python
position1=[]
for i in range(0,10):
  position1.append(position[i].text)

print(position1)
```

After using the soup.findall() we  scrapped the data but it is with the html tag to remove the html tag we have same the data in a new list and in a test format

```python
country1=[]
for i in range(0,10):
  country1.append(country[i].text)

print(country1)
```

```python
players1=[]
for i in range(0,10):
  players1.append(player[i].text)

print(players1)
```

```python
rating=s.find_all('div',class_="cb-col cb-col-17 cb-rank-tbl pull-right")
rating
```

```python
rating1=[]
for i in range(0,10):
  rating1.append(rating[i].text)

print(rating1)
```

```python
df=pd.DataFrame()
df['position']=position1
df["Name"]=players1
df['country']=country1
df['Rating']=rating1
df
```

last step is to convert the data into tabular form for that we will use the pandas library of the python By converting all list data into tabular form

```python
df.to_csv("Icc_Test_Ranking.csv",index=False)
df1=pd.read_csv('Icc_Test_Ranking.csv')
df1
```
